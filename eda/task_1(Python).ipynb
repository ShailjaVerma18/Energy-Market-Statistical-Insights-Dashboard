```python
# import the required library for cleaning the dataset given by Apana_time 
import pandas as pd
```

```python
# Step 1: Load the Excel dataset
df = pd.read_excel("Data set (2).xlsx")
df.head(5)
```

```python
# Step 2: Finding the Missing Value with Percentage as well
missing_values = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100

# Showing both in one table
missing_summary = pd.DataFrame({'Missing Values': missing_values, 'Percentage (%)': missing_percent})
print(missing_summary)
```

```python
# Step 3: Replace all missing (null) values with the mean of that column
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
# Check again if any null values remain
print(df.isnull().sum())
```

```python
# Step 4 : Save the Clean Dataset
df.to_excel("Cleaned_Data.xlsx", index=False)
print("Cleaned file saved successfully!")
```

#### Performing EDA i.e. Exploratory Data Analysis

###### 1. Measures of Central Tendency → Mean, Median, Mode

###### 2. Measures of Dispersion → Variance, Standard Deviation, Range

###### 3. Shape of Distribution → Skewness, Kurtosis

```python
# Read the clean dataset 
clean_data = pd.read_excel("Cleaned_Data.xlsx")
clean_data.head(10)
```

```python
# Step1: Measures of Central Tendency
# 1.1 Mean (average)
mean_values = clean_data.mean(numeric_only=True)
print("Mean:\n", mean_values)
```

```python
# 1.2 Median (Middle Value)
median_values = clean_data.median(numeric_only=True)
print("\nMedian:\n", median_values)
```

```python
# 1.3 Mode (most frequent value)
mode_values = clean_data.mode(numeric_only=True).iloc[0] # Using iloc[0] to select the first row because clean_data.mode() can return multiple mode values
print("\nMode:\n", mode_values)
```

```python
# Step2: Measures of Dispersion
# 2.1 Variance
variance = clean_data.var(numeric_only=True)
print("\nVariance:\n", variance)
```

```python
# 2.2 Standard Deviation
std_dev = clean_data.std(numeric_only=True)
print("\nStandard Deviation:\n", std_dev)
```

```python
# 2.3 Range
range_values = clean_data.max(numeric_only=True) - df.min(numeric_only=True)
print("\nRange:\n", range_values)
```

```python
# Step 3 : Shape of Distribution
# 3.1 Skewness (Skewness tells whether your data leans more towards the left or right side of the average.)
skewness = clean_data.skew(numeric_only=True)
print("\nSkewness:\n", skewness)
```

```python
# 3.2 Kurtosis (Kurtosis measures how tall or flat your data distribution is compared to a normal bell curve.)
kurtosis = clean_data.kurtosis(numeric_only=True)
print("\nKurtosis:\n", kurtosis)
```

```python
# Saving the summary of eda in one excel file
eda_summary = pd.DataFrame({
    'Mean': mean_values,
    'Median': median_values,
    'Mode': mode_values,
    'Variance': variance,
    'Std_Dev': std_dev,
    'Range': range_values,
    'Skewness': skewness,
    'Kurtosis': kurtosis
})

eda_summary.to_excel("EDA_Results.xlsx")
print("EDA results saved successfully!")
```
